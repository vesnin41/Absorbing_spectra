```python
from OptModels import DrudeModel as dm
from materials import Titanium_prop
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ti_blk = dm(wl=np.arange(2e-6, 31e-6, 1e-6),
                T=np.arange(300, 1100, 100),
                metal_props=Titanium_prop,
               )
ti_blk_norm = dm(wl=np.arange(2e-6, 31e-6, 1e-6),
                                T=np.array([300]),
                                metal_props=Titanium_prop,)
```

```python
s = pd.DataFrame()
for i in range(len(ti_blk.wl)):
    index = pd.MultiIndex.from_product([ti_blk.wl[i:i+1], ti_blk.T])
    cond = pd.DataFrame({'C': ti_blk.sigma_0,
                         'Pf': ti_blk.omega_p,
                         'T_relax': ti_blk.tau,
                         'n_1': ti_blk.n_1[i,:],
                         'k_1': ti_blk.k_1[i,:],
                         'Reflectivity': ti_blk.R[i,:],
                         'Asorptivity': ti_blk.A[i,:]}, 
                        index=index)
    s = s.append(cond)


index = pd.MultiIndex.from_tuples(s.index)
conduct = s.reindex(index)
conduct.index.names = ['Wavelength, mkm', 'Temperature, K']
#del conduct[0]
conduct.index
conduct.loc[(10e-6), :]
conduct
```
