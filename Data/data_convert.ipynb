```python
import numpy as np
import pandas as pd
```

```python
Ti_Palik = pd.read_csv('Ti_Palik.csv')
Ti_Rakic_LD_n = pd.read_csv('Ti_Rakic-LD_n.csv')
Ti_Rakic_LD_k = pd.read_csv('Ti_Rakic-LD_k.csv')
Ti_Rakic_BB_n = pd.read_csv('Ti_Rakic-BB_n.csv')
Ti_Rakic_BB_k = pd.read_csv('Ti_Rakic-BB_k.csv')
Ti_Ordal = pd.read_csv('Ordal.csv')
```

```python
Ti_Palik = Ti_Palik.rename({'Wavelength': 'wl'}, axis='columns')
```

```python
Ti_Rakic_LD = pd.merge(Ti_Rakic_LD_n, Ti_Rakic_LD_k, on='wl')
```

```python
Ti_Ordal = pd.merge(Ti_Ordal.iloc[:50], Ti_Ordal.iloc[51:], on='wl')
```

```python
Ti_Ordal = Ti_Ordal.rename({'n_x': 'n', 'n_y': 'k'}, axis='columns')
```

```python
pd.to_numeric(Ti_Ordal['wl'])
```

```python
Ti_Ordal['wl'] = pd.to_numeric(Ti_Ordal['wl']) * 1e-6
Ti_Ordal['n'] = pd.to_numeric(Ti_Ordal['n'])
Ti_Ordal['k'] = pd.to_numeric(Ti_Ordal['k'])
```

```python
Ti_Ordal.to_csv('Ti_Ordal.csv', sep=',')
Ti_Rakic_LD.to_csv('Ti_Rakic_LD.csv', sep=',')
Ti_Rakic_BB.to_csv('Ti_Rakic_BB.csv', sep=',')
```
